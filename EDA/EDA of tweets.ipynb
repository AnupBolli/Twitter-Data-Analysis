```python
import pandas as pd
import numpy as np
from sortedcontainers import SortedDict
import matplotlib
import matplotlib.pyplot as plt
from statistics import mean 
```

```python
All_Tweets=pd.read_csv('/content/drive/My Drive/Final_Practicum/sentiment140/AllTweets.csv')
```

```python
length = All_Tweets['text'].str.split().str.len()
print(length)
```

```python
freq = {}
for item in length:
    if item>0:
        item=int(item)
        freq[item] = freq.get(item, 0) + 1
    
freq = SortedDict(freq)
print(freq)
```

```python
fig, ax = plt.subplots()
ax.plot(freq.keys(), freq.values())

ax.set(xlabel='Number of words in a tweet', ylabel='Number of tweets',
       title='Distribution of length of tweets')
ax.grid()

fig.savefig("histogram.png")
plt.show()
```

```python
cleanedList = [x for x in length if str(x) != 'nan']
numbers = [ int(x) for x in cleanedList ]

average=mean(numbers)

fig1, ax1 = plt.subplots()
ax1.grid()
ax1.set_title('Distribution of length of tweets')

fig1.savefig("boxplot.png")
ax1.boxplot(numbers)

print(average)
```
